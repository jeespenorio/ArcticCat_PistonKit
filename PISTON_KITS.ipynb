```python
import pandas as pd

# Define columns for the DataFrame
columns = ['MAKE', 'MODEL', 'YEAR', 'VENDOR', 'VENDOR NOTES', 'AD ITEM', 'NI#', 'OEM']

# Create an empty DataFrame with defined columns
df = pd.DataFrame(columns=columns)

# Sample data to add for KITTY
data1_kitty = [
    {
        'MAKE': 'ARCTIC CAT',
        'MODEL': 'KITTY CATS',
        'YEAR': '',
        'VENDOR': 'NACH',
        'NI#': 'SM-09009N, SM-09009R',
        'VENDOR NOTES': '41.00mm Piston Kits , 41.00mm Piston Ring',
        'AD ITEM': '9690PS',
        'OEM': '3002-476'
    }
]

# Append data for the range of years
for year in range(1976, 1991):
    for data in data1_kitty:
        data_copy = data.copy()
        data_copy['YEAR'] = year
        ni_values = data_copy['NI#'].split(', ')  # Split 'NI#' values
        
        for ni_value in ni_values:
            ni_value = ni_value.strip()  # Assign each separated 'NI#' value
            data_copy['NI#'] = ni_value
            
            if ni_value == 'SM-09009N':
                data_copy['VENDOR NOTES'] = '41.00mm Piston Kits'
            elif ni_value == 'SM-09009R':
                data_copy['VENDOR NOTES'] = '41.00mm Piston Ring'

            df = pd.concat([df, pd.DataFrame([data_copy])], ignore_index=True)  # Concatenate DataFrames

# Group by 'VENDOR NOTES' and 'YEAR' to consolidate the values
grouped_df = df.groupby(['VENDOR NOTES', 'YEAR']).agg({
    'MAKE': 'first',
    'MODEL': 'first',
    'VENDOR': 'first',
    'NI#': ','.join,
    'AD ITEM': 'first',
    'OEM': 'first'
}).reset_index()

print(grouped_df)
```
